```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_wine
wine = load_wine()
```

```python
print("Keys Wine:\n{}".format( wine.keys()))
```

```python
print("Первые 15 строк массива data: \n{}".format(wine['data'][:5]))
```

```python
print(wine['target'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(wine['data'], wine['target'], random_state=0)
```

```python
print("Форма массива X_train:{}".format(X_train.shape))
print("Форма массива y_train:{}".format(y_train.shape))
```

```python
print("Форма массива X_test:{}".format(X_test.shape))
print("Форма массива y_test:{}".format(y_test.shape))
```

```python
wine_dataframe = pd.DataFrame(X_train, columns = wine.feature_names)
```

```python
grr = pd.plotting.scatter_matrix(wine_dataframe, c=y_train, figsize=(25, 25), marker='o', hist_kwds={'bins': 20}, s=60, alpha=.8)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=22)
```

```python
knn.fit(X_train, y_train)
```

```python
X_new = np.array([[13.5, 2.5, 2.3, 19, 137, 2.7, 2.8, 0.4, 1.3, 5.0, 1.1, 3.2, 760]])
print("Форма массива X_new: {}".format(X_new.shape))
```

```python
prediction = knn.predict(X_new)
print("Прогноз: {}".format(prediction))
print("Спрогнозированнная метка: {}".format(wine['target_names'][prediction]))
```

```python
y_pred = knn.predict(X_test)
print("Прогнозы для тестового набора:\n {}".format(y_pred))
```

```python
print("Правильность на тестовом наборе: {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
print("Правильность на тестовом наборе: {:.2f}".format(knn.score(X_test, y_test)))
```
